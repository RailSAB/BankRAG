```python
def embed_query(text: str) -> list[float]:
        res = requests.request(method="POST", url=f"http://localhost:8081/vectors", json={"text": text}, headers={"Content-Type": "application/json"})
        res.raise_for_status()
        return res.json()["vector"]
```

```python
embed_query("Hello world")
```
